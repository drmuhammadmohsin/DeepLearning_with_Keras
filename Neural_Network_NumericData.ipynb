# Data Preparation and Data Preprocessing

```python
import numpy as np
from random import randint
from sklearn.utils import shuffle
from sklearn.preprocessing import MinMaxScaler
```

```python
train_labels = []
train_samples = []
```

```python
for i in range(50):
  random_younger = randint(13,64)
  train_samples.append(random_younger)
  train_labels.append(1)
  random_older = randint(65,100)
  train_samples.append(random_older)
  train_labels.append(0)

for i in range(1000):
  random_younger = randint(13,64)
  train_samples.append(random_younger)
  train_labels.append(0)
  random_older = randint(65,100)
  train_samples.append(random_older)
  train_labels.append(1)
```

```python
for i in train_samples:
  print(i)
```

```python
for i in train_labels:
  print(i)
```

```python
train_samples = np.array(train_samples)
train_labels =  np.array(train_labels)
train_samples, train_labels = shuffle(train_samples, train_labels)
```

```python
scaler =  MinMaxScaler( feature_range=(0,1))
scaled_train_sample =  scaler.fit_transform(train_samples.reshape(-1,1))
for i in scaled_train_sample:
  print(i)
```

## Test Data Generation and Preprocessing


```python
test_labels = []
test_samples = []
```

```python
for i in range(50):
  random_younger = randint(13,64)
  test_samples.append(random_younger)
  test_labels.append(1)
  random_older = randint(65,100)
  test_samples.append(random_older)
  test_labels.append(0)

for i in range(1000):
  random_younger = randint(13,64)
  test_samples.append(random_younger)
  test_labels.append(0)
  random_older = randint(65,100)
  test_samples.append(random_older)
  test_labels.append(1)
```

```python
test_samples = np.array(test_samples)
test_labels =  np.array(test_labels)
test_samples, test_labels = shuffle(test_samples, test_labels)
```

```python
scaler =  MinMaxScaler( feature_range=(0,1))
scaled_test_sample =  scaler.fit_transform(test_samples.reshape(-1,1))
for i in scaled_test_sample:
  print(i)
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Activation, Dense

from tensorflow.keras.optimizers import Adam
from tensorflow.keras.metrics import CategoricalCrossentropy
```

```python
model = Sequential([

                    Dense(units=16, input_shape=(1,), activation='relu' ),
                    Dense(units=32,  activation='relu' ),
                    Dense(units=2, activation='softmax' )

])
```

```python
model.summary()
```

```python
model.compile (optimizer=Adam(learning_rate=0.001), loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
history =model.fit(x=scaled_train_sample, y=train_labels, batch_size=10, epochs=30, shuffle=True, verbose=2)
```

```python
# list all data in history
import matplotlib.pyplot as plt
print(history.history.keys())
# summarize history for accuracy
plt.plot(history.history['accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='lower right')
plt.show()
```

```python
plt.plot(history.history['loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train_loss', 'test'], loc='upper right')
plt.show()
```

## how to build validation set

```python
history =model.fit(x=scaled_train_sample, y=train_labels, validation_split=0.1, batch_size=10, epochs=30, shuffle=True, verbose=2)
```

```python
# summarize history for accuracy
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
# summarize history for loss
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```

```python
predictions = model.predict(x=test_samples, batch_size=10, verbose=0)
```

```python
for i in predictions:
  print(i)
```

```python
rounded_predictions = np.argmax(predictions, axis=-1)
for i in rounded_predictions:
  print(i)
```
